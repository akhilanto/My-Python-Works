```python
from selenium import webdriver 
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC 
from selenium.common.exceptions import TimeoutException
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\akhil\Desktop\Research\Phase 2\ieeelinks.csv")
```

```python
option = webdriver.ChromeOptions()
option.add_argument(" â€” incognito")
```

```python
browser = webdriver.Chrome(executable_path="c:/chromedriver/chromedriver.exe")
webdriver_obj = browser
```

```python
Titles =[]
Abstracts =[]
missedlinkis =[]
def exract(url):
    link = url
    print(url)
    try :
        webdriver_obj.get(link)
        titles_element = webdriver_obj.find_element_by_class_name("document-title")
        title=titles_element.text
        Titles.append(title)
        Abstract_element = webdriver_obj.find_element_by_class_name("document-abstract")
        abstract =Abstract_element.text
        Abstracts.append(abstract)
    except:
        print('acm')
        missedlinkis.append(url)
```

```python
Titles =[]
Abstracts =[]
for link in df['doilink'][12315:len(df)]:
    exract(link)
```

```python
doi = pd.DataFrame ( { 'Titles' : Titles ,'Abstracts' : Abstracts})
doi.to_csv('abstract2.csv')
```

```python
missedlinkis = pd.DataFrame ( { 'missedlinkis' : missedlinkis})
missedlinkis.to_csv('missedlinks2.csv')
```
