```python
df3 = dums.drop('encounter_id',1)
```

```python
feat = df3.drop('readmitted',1)
label = df3.readmitted
```

```python
feat.shape
```

```python
label.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtr,xts,ytr,yts = train_test_split(feat,label)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(xtr,ytr)
```

```python
y_pred = log.predict(xts)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,yts)
```

```python
for i in [0.01,0.1,0.4,0.7,1]:
    log = LogisticRegression(C=i)
    log.fit(xtr,ytr)
    y_pred = log.predict(xts)
    print(accuracy_score(y_pred,yts)) 
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
rf.fit(xtr,ytr)
y_pred = rf.predict(xts)
print(accuracy_score(y_pred,yts)) 
```
